```python
files = glob.glob('../../metrics/epoch_metrics/*synchronous*')
```

```python
files[:2]
```

```python
for file in files:
    df = pd.read_parquet(file)
    
    if df['batch_size'].unique() == [32]:
        df = df[df.epoch <= 100]
        df.to_parquet(file)
    elif df['batch_size'].unique() == [64]:
        df = df[df.epoch <= 200]
        df.to_parquet(file)
    elif df['batch_size'].unique() == [128]:
        df = df[df.epoch <= 400]
        df.to_parquet(file)
    elif df['batch_size'].unique() == [256]:
        df = df[df.epoch <= 800]
        df.to_parquet(file)
```
